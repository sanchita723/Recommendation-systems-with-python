```python
credits_rename = credits.rename(columns={'movie_id':'id'},index=str) 
merge_df= movies_df.merge(credits_rename, on='id')
merge_df.head()
```

```python
merge_df.columns
```

```python
movies_clean_df=merge_df.drop(columns=['homepage','title_x','title_y','status','production_countries'])
movies_clean_df.head()
```

```python
movies_clean_df.info()
```

```python
#weighted average technique
v=movies_clean_df['vote_count']
r=movies_clean_df['vote_average']
c=movies_clean_df['vote_average'].mean()
m=movies_clean_df['vote_count'].quantile(0.70)
```

```python
movies_clean_df['weighted_avg']=((r*v)+(c*m))/(v+m)
movies_clean_df.head()
```

```python
movies_sorted_ranking=movies_clean_df.sort_values('weighted_avg',ascending=False)
movies_sorted_ranking[['original_title','vote_count','vote_average','weighted_avg','popularity']].head(20)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.barplot(x='vote_average',y='original_title',data=movies_sorted_ranking.head(10))
plt.title('Best Movies By Voting Avg', weight='bold')
plt.figure(figsize=(15,7))
```

```python
popularity=movies_clean_df.sort_values('popularity',ascending=False)
sns.barplot(x='popularity',y='original_title',data=popularity.head(10))
plt.title('Best Movies By Popularity', weight='bold')
plt.figure(figsize=(15,7))
```

### Recommendation system based on 50% weighted avg and 50% popularity

```python
from sklearn.preprocessing import MinMaxScaler

scaling=MinMaxScaler()
movie_scaled_df=scaling.fit_transform(movies_clean_df[['weighted_avg','popularity']])
df=pd.DataFrame(movie_scaled_df,columns=['weighted_avg','popularity'])
df.head()
```

```python
movies_clean_df[['normalized_weighted_avg','normalized_weighted_popularity']]=df
movies_clean_df.head()
```

```python
movies_clean_df['score']=movies_clean_df['normalized_weighted_avg']*0.5 + movies_clean_df['normalized_weighted_popularity']*0.5
movies_scored_df=movies_clean_df.sort_values(['score'],ascending=False)
movies_scored_df[['original_title','normalized_weighted_avg','normalized_weighted_popularity','score']].head(10)
```

```python
sns.barplot(x='score',y='original_title',data=movies_scored_df.head(10))
plt.title('Best Movies By Popularity And Avg Weighting', weight='bold')
plt.figure(figsize=(15,7))
```
